```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
```

```python
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()

# Function to calculate sentiment
def get_sentiment(text):
    score = sia.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'Positive'
    elif score['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment']= df['text'].apply(get_sentiment)
```

```python
x= df['text']
y= df['Sentiment']
```

```python
print(f"Original x shape: {len(x)}") 
print(f"Original y shape: {len(y)}")
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,test_size=0.2,stratify=y,shuffle=True)
```

```python
print("shape of x_test=", x_test.shape)
print("shape of x_train=", x_train.shape)
print("shape of y_test=", y_test.shape)
print("shape of y_train=", y_train.shape)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, LSTM, GRU
```

```python
y
```

```python
max_vocab_size= 20000
max_sequence_length= 100
```

```python
# Initialize and fit tokenizer
tokenizer= Tokenizer(num_words=max_vocab_size)
tokenizer.fit_on_texts(x_train)
```

```python
tokenizer
```

# Convert the text data into sequences

```python
x_train_seq= tokenizer.texts_to_sequences(x_train)
```

```python
x_test_seq= tokenizer.texts_to_sequences(x_test)
```

```python
x_train[0]
```

```python
x_train_seq[0]
```

```python
x_train_seq
```

# Pad the sequences

```python
x_train_pad = pad_sequences(x_train_seq, maxlen=max_sequence_length)
x_test_pad = pad_sequences(x_test_seq,maxlen=max_sequence_length)
```

```python
print(f"x_train_pad shape: {x_train_pad.shape}")  # Expected: (398, 100)
print(f"x_test_pad shape: {x_test_pad.shape}") 
```

```python
x_train_pad[0]
```

```python
embedding_dim=100
```

```python
from keras.layers import Dense,BatchNormalization
from keras.layers import Flatten
```

```python
model= Sequential()
model.add(Embedding(input_dim=max_vocab_size,output_dim=embedding_dim,input_length=max_sequence_length))
model.add(LSTM(units=64,return_sequences=True, activation='relu'))
model.add(BatchNormalization())
model.add(LSTM(units=32,return_sequences=True, activation='relu'))
model.add(BatchNormalization())
model.add(LSTM(units=16,return_sequences=True, activation='relu'))
model.add(BatchNormalization())
model.add(LSTM(units=8,return_sequences=False, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss= 'categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train_pad,y_train,batch_size=32,epochs=30,validation_split=0.2)
```

```python
print(x_test_pad.shape)
print(y_test.shape)
```

```python
test_loss, test_accuracy = model.evaluate(x_test_pad, y_test)
print(f'Test Accuracy: {test_accuracy}')
```

```python
def preprocess_text(text, tokenizer, max_sequence_length):
    text_seq = tokenizer.texts_to_sequences([text])
    text_pad = pad_sequences(text_seq, maxlen=max_sequence_length)
    return text_pad
```
